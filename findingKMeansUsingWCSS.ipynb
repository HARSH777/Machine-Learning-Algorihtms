```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as py
```

```python
dataset = pd.read_csv('Customers.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,[2,3]].values
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=10)
```

```python
y_mean =km.fit_predict(X)
```

```python
py.scatter(X[y_mean==0,0], X[y_mean==0,1],s=100,c='blue',label='Cluster1')
py.scatter(X[y_mean==1,0], X[y_mean==1,1], s= 100, c='red', label ='cluster2')
py.scatter(X[y_mean==2,0], X[y_mean==2,1], s= 100, c='orange', label ='cluster2')
py.scatter(X[y_mean==3,0], X[y_mean==3,1], s= 100, c='green', label ='cluster2')
py.scatter(X[y_mean==4,0], X[y_mean==4,1], s= 100, c='magenta', label ='cluster2')
py.scatter(X[y_mean==5,0], X[y_mean==5,1], s= 100, c='yellow', label ='cluster2')
py.scatter(X[y_mean==6,0], X[y_mean==6,1], s= 100, c='gray', label ='cluster2')
py.scatter(X[y_mean==7,0], X[y_mean==7,1], s= 100, c='black', label ='cluster2')
py.scatter(X[y_mean==8,0], X[y_mean==8,1], s= 100, c='purple', label ='cluster2')
py.scatter(X[y_mean==9,0], X[y_mean==9,1], s= 100, c='brown', label ='cluster2')
py.scatter(X[y_mean==10,0], X[y_mean==10,1], s= 100, c='silver', label ='cluster2')
py.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],c='yellow', s = 300)
```

```python
wcss=[]
```

```python
for i in range(1,15):
    km=KMeans(n_clusters=i)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
py.figure(figsize=[12.0, 10.0])
py.plot(range(1,15),wcss)
py.xlabel('number of cluster')
py.ylabel('wcss')
py.title('within cluster sum of squares')
```
