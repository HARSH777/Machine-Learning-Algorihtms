```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as py
```

```python
dataset  = pd.read_csv('Customers.csv')
```

```python
X=dataset.iloc[:,[2,3]].values
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
```

```python
for i in range(1,10):
    km = KMeans(n_clusters=i)
    km.fit(X)
    wcss.append(km.inertia_)
```

```python
py.plot(range(1,10), wcss)
```

```python
km.fit(X)
```

```python
kmean = km.predict(X)
```

```python
py.scatter(X[kmean==0,0], X[kmean==0,1], s=100, c='blue')
py.scatter(X[kmean==1,0], X[kmean==1,1], s=100, c='green')
py.scatter(X[kmean==2,0], X[kmean==2,1], s=100, c='red')
py.scatter(X[kmean==3,0], X[kmean==3,1], s=100, c='black')
py.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1], s= 200, c ='yellow')
```
